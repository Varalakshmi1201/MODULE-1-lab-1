Data visualization:Data visualization:
Data visualization is the process of
Creating graphical representations of
data to better understand and communi-

```python
import matplotlib.pyplot as plt

# sample data
X = [8,6,11,15,5]
Y= [16,12,9,20,14]

# Create a line plot
plt.plot(X,Y,marker='*',color='r')
#Add titles and labels
plt.title('title Representation ')
plt.xlabel('X-axis')
plt.ylabel('y-axis')

#Display the plot
plt.show()
```

```python

Bar Graphs:
A bar graph (or bar chart)is a
Graphical representation of data that
Uses rectangular bars to display the
Magnitude or frequency of different
categories
```

```python
import matplotlib.pyplot as plt

# Data
months = ['April', 'May', 'Jun', 'Jul', 'May']
sales = [100, 120, 150, 180, 200]

# Create bar graph
plt.bar(months, sales)

# Set title and labels
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')

# Show plot
plt.show()
```

Pie Chart:

A pie chart is a circular graphical representation of data that shows how different categories contribute to a whole.

```python
import matplotlib.pyplot as plt

# Data
months = ['Apr', 'May', 'Jun', 'Jul', 'May']
sales = [100, 120, 150, 180, 200]
plt.pie(sales,labels=months,autopct='%2.2f%%',colors=['green','Pink','violet','red','blue'])
plt.title('Monthly Sales')
plt.xlabel('Months')
plt.ylabel('Sales')

plt.show()
```
